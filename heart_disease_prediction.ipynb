# Heart Disease Prediction

This notebook shows how we can predict if a person has heart disease using machine learning.  
We use a dataset with patient details like age, cholesterol, blood pressure, chest pain type, and more.  
The goal is to train a model that can identify heart disease early and help in better healthcare decisions.

---

## Steps in the Project

### 1. Data Loading and Checking
- Load the dataset
- Look at the first few rows
- Check data types, summary, and missing values

### 2. Data Preparation
- Make a copy of the dataset
- Check the balance of target values (heart disease vs no heart disease)
- Fix imbalance so the model learns from both classes fairly

### 3. Model Training
- Use Logistic Regression (baseline model)
- Use Support Vector Machine (SVM) for better predictions
- Split the data into training and test sets
- Train the models and make predictions

### 4. Model Evaluation
- Check Accuracy (how often the model is correct)
- Check Precision (how many predicted positives are correct)
- Check Recall (how many actual positives are found)
- Use F1-score to balance precision and recall

### 5. Conclusion
- Balanced data improves results
- Logistic Regression works as a simple baseline
- SVM gives good accuracy on this dataset
- Machine learning can be useful in detecting heart disease early


```python


import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
loan_data = pd.read_csv('heart_disease_data.csv')
loan_data.head()
```

```python
loan_data.tail()
```

```python
loan_data.info()
```

```python
loan_data.describe()
```

```python
loan_data.isnull().sum()
```

```python
loan_data['target'].value_counts()
```

```python
new_data = loan_data.copy()
```

```python
legit = new_data[new_data['target']== 1]
fraud = new_data[new_data['target']== 0]
```

```python
legit_sample = legit.sample(n=138)
```

```python
data = pd.concat([legit_sample,fraud],axis = 0)
```

```python
data['target'].value_counts()
```

```python
data.head()
```

```python
x = data.drop(columns = ['target','slope'],axis = 1)
y = data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
model = LogisticRegression(max_iter = 10000)
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_prediction,y_train)
print('training accuracy :', x_train_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction,y_test)
print('testing accuracy :', x_test_accuracy)
```

```python
input_data = (41,0,1,130,204,0,0,172,0,1.4,0,2)

input_data_np = np.asarray(input_data)
reshaped_data = input_data_np.reshape(1,-1)
```

```python
prediction = model.predict(reshaped_data)
print('prediction :', prediction)
```

```python
if(prediction == [1]):
    print('Likely to have heart disease')

else:
    print('Not likely to have heart disease')
```
